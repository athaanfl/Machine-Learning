```python
#@title Sel 1: Hubungkan ke Google Drive
# Jalankan sel ini untuk menghubungkan notebook dengan Google Drive Anda.
import os
from google.colab import drive

print("Menghubungkan ke Google Drive...")
try:
    drive.mount('/content/drive')
    print("Google Drive berhasil terhubung.")
except Exception as e:
    print(f"Gagal menghubungkan Google Drive: {e}")

#@title Sel 2: Ekstrak File ZIP dari Drive
import os

# --- Silakan Ganti Bagian Ini ---
# 1. Ganti dengan path LENGKAP ke file .zip Anda di Google Drive.
#    Klik kanan pada file zip di Drive, pilih "Copy path", lalu tempel di sini.
PATH_TO_ZIP_FILE = "/content/drive/MyDrive/oot/FishImgDataset.zip"

# 2. Tentukan di mana Anda ingin mengekstraknya di Colab (biasanya /content/ sudah cukup).
DESTINATION_DIR = "/content/"
# --- Batas Modifikasi ---

# Mengekstrak file .zip
if os.path.exists(PATH_TO_ZIP_FILE):
    print(f"\nMengekstrak {os.path.basename(PATH_TO_ZIP_FILE)}...")
    !unzip -q "{PATH_TO_ZIP_FILE}" -d "{DESTINATION_DIR}"
    print("Ekstraksi selesai.")
else:
    print(f"!!! Peringatan: File ZIP di '{PATH_TO_ZIP_FILE}' tidak ditemukan.")
    print("!!! Pastikan path sudah benar.")

#@title Sel 3: Pemuatan Data, Pelatihan, dan Evaluasi Model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, RandomFlip, RandomRotation, RandomZoom, RandomContrast, Rescaling
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Langkah 1: Pengaturan Path dan Pemuatan Data ---
print("\n--- Memulai Pengaturan Path dan Pemuatan Data ---")

# PERBAIKAN: Path sekarang mengarah ke folder hasil ekstraksi di Colab, bukan di Drive.
# Ini akan membuat proses pemuatan data jauh lebih cepat.
# Ganti 'FishImgDataset' jika nama folder utama di dalam file zip Anda berbeda.
base_dataset_path = '/content/FishImgDataset'

TRAIN_DIR = os.path.join(base_dataset_path, 'train')
TEST_DIR = os.path.join(base_dataset_path, 'test')
VAL_DIR = os.path.join(base_dataset_path, 'val')

# Memeriksa apakah path benar-benar ada
if not os.path.exists(base_dataset_path):
    print(f"!!! PERINGATAN: Path dasar '{base_dataset_path}' tidak ditemukan setelah ekstraksi.")
    print("!!! Pastikan nama folder di variabel 'base_dataset_path' sudah sesuai dengan nama folder di dalam .zip Anda.")
else:
    print(f"Path folder training: {TRAIN_DIR}")
    print(f"Path folder testing: {TEST_DIR}")
    print(f"Path folder validation: {VAL_DIR}")


    # Parameter
    IMG_SIZE = (128, 128)
    BATCH_SIZE = 32
    EPOCHS = 25 # Anda bisa menaikkan ini jika hasilnya belum maksimal, misal: 50

    # Fungsi untuk memuat data dari direktori
    def load_data(directory, subset_name):
        print(f"Memuat data dari: {directory}")
        try:
            dataset = tf.keras.utils.image_dataset_from_directory(
                directory,
                label_mode='categorical', # Menggunakan 'categorical' untuk one-hot encoding
                image_size=IMG_SIZE,
                batch_size=BATCH_SIZE,
                shuffle=(subset_name == 'training') # Hanya shuffle data training
            )
            return dataset
        except Exception as e:
            print(f"Gagal memuat data dari {directory}. Error: {e}")
            print("Pastikan struktur folder benar: train, test, dan val berisi subfolder nama-nama ikan.")
            return None

    train_dataset = load_data(TRAIN_DIR, "training")
    val_dataset = load_data(VAL_DIR, "validation")
    test_dataset = load_data(TEST_DIR, "testing")

    # Hentikan jika salah satu dataset gagal dimuat
    if not all([train_dataset, val_dataset, test_dataset]):
        print("\nEksekusi dihentikan karena satu atau lebih dataset gagal dimuat.")
    else:
        class_names = train_dataset.class_names
        NUM_CLASSES = len(class_names)
        print(f"\nDataset berhasil dimuat. Ditemukan {NUM_CLASSES} kelas.")

        # --- Langkah 2: Augmentasi Data & Preprocessing ---
        print("\n--- Menyiapkan Augmentasi dan Preprocessing Data ---")

        data_augmentation = Sequential([
            RandomFlip("horizontal"),
            RandomRotation(0.2),
            RandomZoom(0.2),
            RandomContrast(0.2),
        ], name="data_augmentation")

        AUTOTUNE = tf.data.AUTOTUNE
        def preprocess_dataset(dataset, is_training=False):
            ds = dataset.map(lambda x, y: (Rescaling(1./255)(x), y), num_parallel_calls=AUTOTUNE)
            if is_training:
                ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)
            return ds.prefetch(buffer_size=AUTOTUNE)

        train_ds = preprocess_dataset(train_dataset, is_training=True)
        val_ds = preprocess_dataset(val_dataset)
        test_ds = preprocess_dataset(test_dataset)
        print("Preprocessing selesai.")

        # --- Langkah 3: Membangun Model CNN ---
        print("\n--- Membangun Arsitektur Model CNN ---")

        model = Sequential([
            Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3)),
            Conv2D(32, (3, 3), activation='relu', padding='same'),
            MaxPooling2D((2, 2)),
            Conv2D(64, (3, 3), activation='relu', padding='same'),
            MaxPooling2D((2, 2)),
            Conv2D(128, (3, 3), activation='relu', padding='same'),
            MaxPooling2D((2, 2)),
            Conv2D(256, (3, 3), activation='relu', padding='same'),
            MaxPooling2D((2, 2)),
            Flatten(),
            Dense(512, activation='relu'),
            Dropout(0.5),
            Dense(NUM_CLASSES, activation='softmax')
        ], name="Fish_Classifier")

        model.compile(
            optimizer=Adam(learning_rate=0.0005),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        model.summary()

        # --- Langkah 4: Melatih Model ---
        print("\n--- Memulai Pelatihan Model ---")
        early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        history = model.fit(
            train_ds,
            epochs=EPOCHS,
            validation_data=val_ds,
            callbacks=[early_stopping],
            verbose=1
        )
        print("Pelatihan selesai.")

        # --- Langkah 5: Evaluasi Model ---
        print("\n--- Memulai Evaluasi Model pada Data Tes ---")
        loss, accuracy = model.evaluate(test_ds)
        print(f'Test Accuracy: {accuracy:.4f}')
        print(f'Test Loss: {loss:.4f}')

        print("\nMembuat prediksi pada data tes...")
        y_pred_proba = model.predict(test_ds)
        y_pred = np.argmax(y_pred_proba, axis=1)
        y_true = []
        for images, labels in test_ds:
            y_true.extend(np.argmax(labels.numpy(), axis=1))
        y_true = np.array(y_true)

        print("\nLaporan Klasifikasi (Classification Report):")
        labels_for_report = np.arange(len(class_names))
        print(classification_report(y_true, y_pred, target_names=class_names, labels=labels_for_report, zero_division=0))

        print("\nMembuat Visualisasi Confusion Matrix...")
        cm = confusion_matrix(y_true, y_pred, labels=labels_for_report)
        plt.figure(figsize=(12, 10))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
        plt.title('Confusion Matrix', fontsize=16)
        plt.ylabel('Kelas Aktual (Actual Class)', fontsize=12)
        plt.xlabel('Kelas Prediksi (Predicted Class)', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

        print("\nMembuat Visualisasi Grafik Akurasi dan Loss...")
        acc = history.history['accuracy']
        val_acc = history.history['val_accuracy']
        loss = history.history['loss']
        val_loss = history.history['val_loss']
        epochs_range = range(len(acc))

        plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1)
        plt.plot(epochs_range, acc, label='Training Accuracy')
        plt.plot(epochs_range, val_acc, label='Validation Accuracy')
        plt.legend(loc='lower right')
        plt.title('Training and Validation Accuracy')

        plt.subplot(1, 2, 2)
        plt.plot(epochs_range, loss, label='Training Loss')
        plt.plot(epochs_range, val_loss, label='Validation Loss')
        plt.legend(loc='upper right')
        plt.title('Training and Validation Loss')
        plt.show()
```

3.  **Augmentasi Data Bertarget:** Terapkan teknik augmentasi data yang lebih agresif (misalnya perubahan warna, kecerahan, atau penambahan noise) secara spesifik pada gambar-gambar dari kelas berperforma rendah untuk meningkatkan ketahanan model.

4.  **Implementasi _Transfer Learning_**: Meskipun model saat ini sudah baik, penggunaan arsitektur pra-terlatih seperti **MobileNetV2** atau **EfficientNet** tetap menjadi praktik terbaik.
