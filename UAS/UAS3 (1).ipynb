```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, RandomFlip, RandomRotation, RandomZoom, RandomContrast, Rescaling
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from google.colab import drive

# --- Langkah 0: Hubungkan ke Google Drive ---
# Pastikan Anda mengizinkan akses saat diminta.
print("Menghubungkan ke Google Drive...")
try:
    drive.mount('/content/drive')
    print("Google Drive berhasil terhubung.")
except Exception as e:
    print(f"Gagal menghubungkan Google Drive: {e}")
    # Hentikan eksekusi jika drive tidak terhubung
    exit()

# --- Langkah 1: Pengaturan Path dan Pemuatan Data ---
print("\n--- Memulai Pengaturan Path dan Pemuatan Data ---")

# Ganti dengan path ke direktori dataset Anda di Google Drive
# PASTIKAN PATH INI SUDAH BENAR SESUAI DENGAN LOKASI FOLDER ANDA
base_dataset_path = '/content/drive/MyDrive/oot/Types_of_fish/Types_of_fish/FishImgDataset'

TRAIN_DIR = os.path.join(base_dataset_path, 'train')
TEST_DIR = os.path.join(base_dataset_path, 'test')
VAL_DIR = os.path.join(base_dataset_path, 'val')

# Memeriksa apakah path benar-benar ada
if not os.path.exists(base_dataset_path):
    print(f"!!! PERINGATAN: Path dasar '{base_dataset_path}' tidak ditemukan.")
    print("!!! Pastikan path sudah benar dan folder sudah ada di Google Drive Anda.")
    exit()

print(f"Path folder training: {TRAIN_DIR}")
print(f"Path folder testing: {TEST_DIR}")
print(f"Path folder validation: {VAL_DIR}")


# Parameter
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 25 # Anda bisa menaikkan ini jika hasilnya belum maksimal, misal: 50

# Fungsi untuk memuat data dari direktori
def load_data(directory, subset_name):
    print(f"Memuat data dari: {directory}")
    try:
        dataset = tf.keras.utils.image_dataset_from_directory(
            directory,
            label_mode='categorical', # Menggunakan 'categorical' untuk one-hot encoding
            image_size=IMG_SIZE,
            batch_size=BATCH_SIZE,
            shuffle=(subset_name == 'training') # Hanya shuffle data training
        )
        return dataset
    except Exception as e:
        print(f"Gagal memuat data dari {directory}. Error: {e}")
        print("Pastikan struktur folder benar: train, test, dan val berisi subfolder nama-nama ikan.")
        return None

train_dataset = load_data(TRAIN_DIR, "training")
val_dataset = load_data(VAL_DIR, "validation")
test_dataset = load_data(TEST_DIR, "testing")

# Hentikan jika salah satu dataset gagal dimuat
if not all([train_dataset, val_dataset, test_dataset]):
    print("\nEksekusi dihentikan karena satu atau lebih dataset gagal dimuat.")
else:
    class_names = train_dataset.class_names
    NUM_CLASSES = len(class_names)
    print(f"\nDataset berhasil dimuat. Ditemukan {NUM_CLASSES} kelas: {class_names}")

    # --- Langkah 2: Augmentasi Data & Preprocessing ---
    print("\n--- Menyiapkan Augmentasi dan Preprocessing Data ---")

    data_augmentation = Sequential([
        # Input layer tidak diperlukan di dalam model augmentasi
        RandomFlip("horizontal"),
        RandomRotation(0.2), # Tingkatkan sedikit rotasi
        RandomZoom(0.2),
        RandomContrast(0.2),
    ], name="data_augmentation")

    # Optimalkan pipeline data
    AUTOTUNE = tf.data.AUTOTUNE

    def preprocess_dataset(dataset):
        # Terapkan augmentasi hanya pada data training
        if "train" in dataset.file_paths[0]:
             dataset = dataset.map(lambda x, y: (data_augmentation(x), y), num_parallel_calls=AUTOTUNE)
        # Terapkan normalisasi (Rescaling) ke semua dataset
        dataset = dataset.map(lambda x, y: (Rescaling(1./255)(x), y), num_parallel_calls=AUTOTUNE)
        # Prefetch untuk mempercepat pemrosesan
        return dataset.prefetch(buffer_size=AUTOTUNE)

    train_ds = preprocess_dataset(train_dataset)
    val_ds = preprocess_dataset(val_dataset)
    test_ds = preprocess_dataset(test_dataset)
    print("Preprocessing selesai.")

    # --- Langkah 3: Membangun Model CNN ---
    print("\n--- Membangun Arsitektur Model CNN ---")

    model = Sequential([
        Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3)),

        # Blok Konvolusi 1
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),

        # Blok Konvolusi 2
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),

        # Blok Konvolusi 3
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),

        # Blok Konvolusi 4 (tambahan untuk model lebih dalam)
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),

        # Lapisan Klasifikasi
        Flatten(),
        Dense(512, activation='relu'), # Tambah neuron
        Dropout(0.5),
        Dense(NUM_CLASSES, activation='softmax') # Lapisan output sesuai jumlah kelas
    ], name="Fish_Classifier")

    # Kompilasi model
    model.compile(
        optimizer=Adam(learning_rate=0.0005), # Kecilkan learning rate sedikit
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    model.summary()

    # --- Langkah 4: Melatih Model ---
    print("\n--- Memulai Pelatihan Model ---")

    # Gunakan EarlyStopping untuk menghentikan training jika tidak ada peningkatan
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    history = model.fit(
        train_ds,
        epochs=EPOCHS,
        validation_data=val_ds,
        callbacks=[early_stopping],
        verbose=1
    )
    print("Pelatihan selesai.")

    # --- Langkah 5: Evaluasi Model ---
    print("\n--- Memulai Evaluasi Model pada Data Tes ---")
    loss, accuracy = model.evaluate(test_ds)
    print(f'Test Accuracy: {accuracy:.4f}')
    print(f'Test Loss: {loss:.4f}')

    # Mendapatkan prediksi dan label asli dari test set
    print("\nMembuat prediksi pada data tes...")
    y_pred_proba = model.predict(test_ds)
    y_pred = np.argmax(y_pred_proba, axis=1)

    # Mengambil label asli dari dataset
    y_true = []
    for images, labels in test_ds:
        y_true.extend(np.argmax(labels.numpy(), axis=1))
    y_true = np.array(y_true)

    # Laporan Klasifikasi (Precision, Recall, F1-Score)
    print("\nLaporan Klasifikasi (Classification Report):")
    # --- PERBAIKAN ---
    # Menambahkan parameter 'labels' untuk memastikan semua kelas (bahkan yang tidak ada di test set) ditampilkan.
    # Ini akan mencegah error jika test set tidak mengandung sampel dari semua kelas.
    labels_for_report = np.arange(len(class_names))
    print(classification_report(y_true, y_pred, target_names=class_names, labels=labels_for_report, zero_division=0))

    # Confusion Matrix
    print("\nMembuat Visualisasi Confusion Matrix...")
    cm = confusion_matrix(y_true, y_pred, labels=labels_for_report)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix', fontsize=16)
    plt.ylabel('Kelas Aktual (Actual Class)', fontsize=12)
    plt.xlabel('Kelas Prediksi (Predicted Class)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Plot Akurasi dan Loss
    print("\nMembuat Visualisasi Grafik Akurasi dan Loss...")
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

```

## II. Analisa Rinci Hasil Eksperimen

### 1. Permasalahan Kritis pada Dataset

Akar dari seluruh permasalahan performa model terletak pada kualitas dan kuantitas dataset.

Kuantitas Data Latih Sangat Kurang: Data latih yang digunakan hanya sebanyak 319 gambar untuk melatih 31 kelas ikan yang berbeda. Ini berarti rata-rata hanya terdapat sekitar 10 gambar per jenis ikan. Jumlah ini secara signifikan tidak memadai bagi model Convolutional Neural Network (CNN) untuk mempelajari fitur-fitur pembeda yang kompleks.

Pembagian Dataset yang Tidak Proporsional:

Set Latih (Train): 319 gambar

Set Validasi (Validation): 2751 gambar

Set Uji (Test): 34 gambar
Distribusi ini tidak wajar dalam praktik machine learning. Set validasi yang ukurannya delapan kali lebih besar dari set latih menyebabkan proses evaluasi selama pelatihan menjadi kurang representatif terhadap kemampuan generalisasi model.

### 2. Indikasi Overfitting Kuat dari Grafik Pelatihan

Grafik akurasi dan loss yang dihasilkan menjadi bukti paling jelas dari fenomena overfitting.

Akurasi Pelatihan (Training Accuracy): Kurva ini menunjukkan tren peningkatan yang konsisten, mencapai sekitar 64%. Ini menandakan model menjadi sangat baik dalam mengenali 319 gambar latih yang sama secara berulang; pada dasarnya, model hanya menghafal data latih.

Akurasi Validasi (Validation Accuracy): Sebaliknya, kurva ini menunjukkan performa yang sangat rendah (tidak pernah melebihi 10%) dan cenderung stagnan.

Kesenjangan Signifikan: Jurang pemisah yang besar antara kurva akurasi pelatihan dan validasi merupakan gejala klasik dari overfitting, di mana model gagal melakukan generalisasi pada data baru.

### 3. Arsitektur Model Terlalu Kompleks untuk Dataset

Meskipun arsitektur CNN yang dibangun adalah standar, model tersebut memiliki 8.7 juta parameter yang perlu dilatih. Upaya untuk melatih model dengan kompleksitas setinggi ini menggunakan dataset yang sangat kecil (319 gambar) secara inheren akan mendorong terjadinya overfitting.

Dari analisa di atas, dapat disimpulkan bahwa performa model yang buruk bukan disebabkan oleh kesalahan pada kode atau arsitektur, melainkan oleh strategi pelatihan yang tidak sesuai dengan kondisi dataset yang ada.

Untuk perbaikan di masa mendatang, disarankan beberapa langkah strategis sebagai berikut:

Mengimplementasikan Transfer Learning: Pendekatan yang paling direkomendasikan adalah menggunakan model pra-terlatih (misalnya MobileNetV2, ResNet50, EfficientNet) yang telah memiliki kemampuan ekstraksi fitur dasar dari jutaan gambar. Proses pelatihan kembali (fine-tuning) hanya pada beberapa lapisan akhir diharapkan dapat meningkatkan akurasi secara drastis dengan data yang terbatas.

Peningkatan Kuantitas dan Keseimbangan Data: Syarat fundamental lainnya adalah penambahan jumlah data latih secara signifikan, dengan target minimal 100 gambar per kelas untuk hasil yang lebih baik. Selain itu, pembagian data perlu diseimbangkan kembali ke proporsi yang lebih wajar (misal: 70% latih, 15% validasi, 15% tes).

Optimalisasi Augmentasi Data: Pemanfaatan teknik augmentasi data akan memberikan dampak yang lebih signifikan jika diterapkan pada dataset latih yang lebih besar dan representatif.
