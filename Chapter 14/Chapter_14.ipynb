```python
# -*- coding: utf-8 -*-
"""
Chapter 14: Deep Computer Vision Using Convolutional Neural Networks - Notebook Reproduksi Kode

Bab ini mengeksplorasi Convolutional Neural Networks (CNNs), arsitektur jaringan saraf
yang sangat efektif untuk tugas-tugas visi komputer.

Kita akan melihat:
- Arsitektur CNN dasar (Convolutional, Pooling, Flatten, Dense).
- Implementasi CNN dengan Keras.
- Data Augmentation.
- Arsitektur CNN terkenal (LeNet-5, AlexNet, GoogLeNet, ResNet) secara konseptual.
"""

import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Atur random seed untuk reproduksibilitas
np.random.seed(42)
tf.random.set_seed(42)

# Path untuk menyimpan gambar plot
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "cnn"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

# Fungsi pembantu untuk plot kurva pembelajaran
def plot_learning_curves(history):
    pd.DataFrame(history.history).plot(figsize=(8, 5))
    plt.grid(True)
    plt.gca().set_ylim(0, 1) # Untuk akurasi/loss
    plt.title("Kurva Pembelajaran")
    plt.xlabel("Epoch")
    plt.ylabel("Metrik")
    plt.show()

# --- 1. Arsitektur CNN Dasar ---

# a. Memuat dan Menyiapkan Dataset Fashion MNIST
print("--- Memuat dan Menyiapkan Dataset Fashion MNIST ---")
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()

# Skala piksel ke rentang 0-1
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0

# Bagi menjadi set pelatihan dan validasi
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]

# CNN mengharapkan input 4D: [batch_size, height, width, channels]
# Fashion MNIST adalah grayscale, jadi tambahkan dimensi channel (1)
X_train = X_train[..., np.newaxis] # Ubah dari (55000, 28, 28) menjadi (55000, 28, 28, 1)
X_valid = X_valid[..., np.newaxis]
X_test = X_test[..., np.newaxis]

print(f"Bentuk X_train setelah penambahan channel: {X_train.shape}")

class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]


# b. Membangun CNN Menggunakan Keras Sequential API
# Arsitektur sederhana: Conv2D -> MaxPooling2D -> Conv2D -> MaxPooling2D -> Flatten -> Dense (Output)
print("\n--- Membangun CNN dengan Sequential API ---")
model_cnn = keras.models.Sequential([
    # Layer Konvolusi 1
    keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu", padding="same",
                        input_shape=[28, 28, 1]), # Input: 28x28 grayscale
    keras.layers.MaxPooling2D(pool_size=(2, 2)), # Ukuran pool 2x2 akan mengurangi ukuran gambar menjadi setengahnya

    # Layer Konvolusi 2
    keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation="relu", padding="same"),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),

    # Flatten untuk terhubung ke Dense layers
    keras.layers.Flatten(),

    # Dense layers (klasifikasi tradisional)
    keras.layers.Dense(units=128, activation="relu"),
    keras.layers.Dense(units=10, activation="softmax") # Output 10 kelas Fashion MNIST
])

print("\n--- Ringkasan Model CNN: ---")
model_cnn.summary()

# c. Mengkompilasi dan Melatih Model CNN
print("\n--- Mengkompilasi dan Melatih Model CNN ---")
model_cnn.compile(loss="sparse_categorical_crossentropy",
                  optimizer="adam", # Adam sering kali pilihan yang baik untuk CNN
                  metrics=["accuracy"])

history_cnn = model_cnn.fit(X_train, y_train, epochs=10,
                            validation_data=(X_valid, y_valid))
print("Model CNN berhasil dilatih.")

# d. Mengevaluasi Model CNN
print("\n--- Mengevaluasi Model CNN pada Set Pengujian ---")
test_loss, test_accuracy = model_cnn.evaluate(X_test, y_test)
print(f"Loss pada set pengujian: {test_loss:.4f}")
print(f"Akurasi pada set pengujian: {test_accuracy:.4f}")

plot_learning_curves(history_cnn)


# --- 2. Data Augmentation ---
# Membangkitkan lebih banyak data pelatihan dari data yang ada
# Contoh: membalik, memutar, menggeser gambar

print("\n--- Data Augmentation ---")

# Data Augmentation menggunakan tf.keras.layers.Random* layers
data_augmentation = keras.models.Sequential([
    keras.layers.RandomFlip("horizontal"), # Membalik gambar secara horizontal
    keras.layers.RandomRotation(0.02), # Memutar gambar kecil-kecilan
    keras.layers.RandomZoom(0.1), # Memperbesar/memperkecil gambar
    # keras.layers.RandomTranslation(height_factor=0.1, width_factor=0.1), # Menggeser gambar
])

# Integrasikan augmentation ke dalam model atau pipeline data
# Option 1: Sebagai bagian dari model Sequential (paling sederhana)
# Letakkan layer augmentation di awal model
model_cnn_aug = keras.models.Sequential([
    data_augmentation, # Layer augmentasi di awal
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=[28, 28, 1]),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model_cnn_aug.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("\nMelatih Model CNN dengan Data Augmentation...")
# Data augmentasi tidak perlu diterapkan pada data validasi/pengujian
history_cnn_aug = model_cnn_aug.fit(X_train, y_train, epochs=10,
                                    validation_data=(X_valid, y_valid))
print("Model CNN dengan Data Augmentation berhasil dilatih.")
print(f"Akurasi pada set pengujian (dengan Augmentation): {model_cnn_aug.evaluate(X_test, y_test, verbose=0)[1]:.4f}")

plot_learning_curves(history_cnn_aug)

# --- 3. Arsitektur CNN Terkenal (Konseptual) ---
# Tidak ada reproduksi kode penuh untuk setiap arsitektur karena kompleksitasnya.
# Penjelasan konsep disajikan dalam komentar.

print("\n--- Arsitektur CNN Terkenal (Konseptual) ---")
print("Arsitektur CNN yang paling penting meliputi:")

print("\n- LeNet-5 (1998):")
print("  Salah satu CNN pertama yang sukses, untuk pengenalan digit tulisan tangan.")
print("  Arsitektur: CONV -> POOL -> CONV -> POOL -> FC -> FC -> Output.")
print("  Menggunakan pooling rata-rata, aktivasi sigmoid/tanh.")
print("  Pentingnya: Menunjukkan efektivitas konvolusi dan pooling.")

print("\n- AlexNet (2012):")
print("  Pemenang ImageNet 2012, menandai dimulainya era Deep Learning.")
print("  Arsitektur lebih besar: 5 lapisan konvolusi, 3 lapisan fully-connected.")
print("  Fitur kunci: ReLU (untuk pertama kalinya), Dropout, Data Augmentation ekstensif, LRN (Local Response Normalization), penggunaan GPU.")

print("\n- GoogLeNet / Inception (2014):")
print("  Pemenang ImageNet 2014. Memperkenalkan modul 'Inception'.")
print("  Ide Inception: Melakukan konvolusi dengan kernel berbeda (1x1, 3x3, 5x5) dan pooling secara paralel pada input yang sama, lalu menggabungkan outputnya.")
print("  Manfaat: Mengoptimalkan penggunaan parameter, lebih dalam tetapi dengan komputasi yang efisien.")

print("\n- ResNet (Residual Network) (2015):")
print("  Pemenang ImageNet 2015. Memperkenalkan 'Residual Blocks'.")
print("  Masalah: Jaringan sangat dalam sulit dilatih karena vanishing gradients dan degradasi kinerja.")
print("  Solusi Residual Block: Menambahkan 'skip connection' atau 'shortcut connection' yang langsung meneruskan input ke output blok, memungkinkan model untuk mempelajari 'residual function' daripada keseluruhan fungsi.")
print("  Manfaat: Memungkinkan pelatihan jaringan yang sangat dalam (100+ lapisan) secara efektif, kinerja yang sangat baik.")

print("\n--- Selesai Reproduksi Kode Chapter 14 ---")
```
