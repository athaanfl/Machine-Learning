seperti clustering, anomaly detection, dan density estimation.

- Aplikasi Clustering (Image Segmentation, Preprocessing, Semi-supervised Learning)
- Gaussian Mixture Models (GMM) dan aplikasinya

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from sklearn.datasets import make_blobs, load_digits, make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, mean_squared_error
from sklearn.mixture import GaussianMixture, BayesianGaussianMixture
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from PIL import Image # FIX: Mengganti matplotlib.image.imread dengan PIL.Image untuk pemuatan gambar yang lebih robust
import mpl_toolkits.mplot3d as Axes3D # Untuk plot 3D jika diperlukan
import matplotlib as mpl # Untuk ListedColormap

# Fungsi pembantu untuk visualisasi
def plot_clusters(X, y=None, labels=None, title="Clusters"):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50)
    plt.title(title)
    plt.xlabel("Fitur 1")
    plt.ylabel("Fitur 2")
    if labels:
        plt.colorbar(ticks=range(len(labels)), label="Label Cluster")
        plt.clim(-0.5, len(labels) - 0.5)
    plt.grid(True)
    plt.show()

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=35, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=2, linewidths=12,
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True, title=""):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    x1, x2 = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[x1.ravel(), x2.ravel()])
    Z = Z.reshape(x1.shape)

    plt.contourf(x1, x2, Z,
                 cmap="Pastel2")
    plt.contour(x1, x2, Z,
                linewidths=1, colors='k')
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title(title)

# --- 1. Clustering ---

# a. K-Means
# Membuat dataset blob sintetis
np.random.seed(42)
X_blobs, y_blobs = make_blobs(n_samples=1000, centers=5, random_state=42)

print("--- K-Means Clustering ---")
print("Data blob sintetis:")
plot_clusters(X_blobs, title="Data Blob Asli (Tanpa Label)")

# Melatih K-Means clusterer
k = 5
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init=10 default di sklearn >= 0.23
print(f"\nMelatih K-Means dengan k={k}...")
kmeans.fit(X_blobs)
print("K-Means berhasil dilatih.")

# Label cluster yang ditetapkan untuk setiap instance
y_pred_kmeans = kmeans.labels_
print(f"Contoh label cluster (5 pertama): {y_pred_kmeans[:5]}")

# Pusat cluster (centroids)
print(f"Pusat cluster (Centroids):\n{kmeans.cluster_centers_}")

# Memprediksi cluster untuk instance baru
X_new_blobs = np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
new_instance_preds = kmeans.predict(X_new_blobs)
print(f"Prediksi cluster untuk instance baru: {new_instance_preds}")

# Memvisualisasikan batas keputusan K-Means (Voronoi tessellation)
plt.figure(figsize=(8, 6))
plot_decision_boundaries(kmeans, X_blobs, title="Batas Keputusan K-Means")
plt.show()

# Jarak ke setiap centroid (soft clustering)
distances_to_centroids = kmeans.transform(X_new_blobs)
print(f"Jarak instance baru ke centroids:\n{distances_to_centroids}")

# b. Kinerja K-Means: Inertia dan Silhouette Score
print("\n--- Evaluasi K-Means: Inertia dan Silhouette Score ---")
# Inertia (sum of squared distances of samples to their closest cluster center)
print(f"Inertia model K-Means (k={k}): {kmeans.inertia_:.2f}")

# Silhouette Score (rata-rata koefisien siluet untuk semua instance)
# Koefisien siluet berkisar -1 hingga +1. Mendekati +1 berarti cluster yang baik.
silhouette_avg = silhouette_score(X_blobs, y_pred_kmeans)
print(f"Silhouette Score model K-Means (k={k}): {silhouette_avg:.4f}")

# Memilih k optimal menggunakan Inertia (Elbow Method) dan Silhouette Score
print("\n--- Memilih k Optimal untuk K-Means ---")
inertias = []
silhouette_scores = []
ks = range(1, 10)

for k_val in ks:
    kmeans_k = KMeans(n_clusters=k_val, random_state=42, n_init=10)
    kmeans_k.fit(X_blobs)
    inertias.append(kmeans_k.inertia_)
    if k_val > 1: # Silhouette score tidak terdefinisi untuk k=1
        silhouette_scores.append(silhouette_score(X_blobs, kmeans_k.labels_))

# Plot Elbow Method
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(ks, inertias, "bo-")
plt.xlabel("k (Jumlah Cluster)")
plt.ylabel("Inertia")
plt.title("Metode Siku (Elbow Method)")
plt.grid(True)

# Plot Silhouette Score
plt.subplot(122)
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("k (Jumlah Cluster)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score")
plt.grid(True)
plt.suptitle("Memilih k Optimal untuk K-Means", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


# c. Mini-batch K-Means (untuk dataset yang sangat besar)
from sklearn.cluster import MiniBatchKMeans
print("\n--- Mini-batch K-Means ---")
minibatch_kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, n_init=10, batch_size=256)
print(f"Melatih Mini-batch K-Means dengan k={k}...")
minibatch_kmeans.fit(X_blobs)
print("Mini-batch K-Means berhasil dilatih.")
print(f"Inertia model Mini-batch K-Means: {minibatch_kmeans.inertia_:.2f}")
print("Catatan: Inertia Mini-batch K-Means biasanya sedikit lebih tinggi dari K-Means biasa.")


# --- 2. Aplikasi Clustering ---

# a. Image Segmentation (Segmentasi Citra)
# Menggunakan K-Means untuk segmentasi warna
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "unsupervised_learning"
IMAGES_PATH_UNSUPERVISED = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH_UNSUPERVISED, exist_ok=True)

# Path gambar (pastikan ada gambar di folder images/unsupervised_learning)
image_path_ladybug = os.path.join("images", "unsupervised_learning", "ladybug.png")
# Buat dummy image jika file tidak ada (untuk memastikan kode tetap berjalan)
if not os.path.exists(image_path_ladybug):
    print(f"File gambar '{image_path_ladybug}' tidak ditemukan. Membuat gambar dummy.")
    dummy_image = np.random.rand(533, 800, 3) # Contoh gambar random
    plt.imsave(image_path_ladybug, dummy_image)

# FIX: Muat gambar menggunakan PIL.Image untuk memastikan format RGB dan kemudian normalisasi.
# Ini menggantikan: image = imread(image_path_ladybug) dan image_np = np.array(image, dtype=np.float32)
try:
    # Memuat gambar menggunakan PIL.Image dan mengkonversinya ke RGB
    pil_image = Image.open(image_path_ladybug).convert('RGB')
    # Mengkonversi ke NumPy array dan normalisasi ke rentang [0, 1]
    image_np = np.array(pil_image, dtype=np.float32) / 255.0
except FileNotFoundError:
    print(f"File gambar '{image_path_ladybug}' tidak ditemukan. Membuat gambar hitam dummy.")
    image_np = np.zeros((533, 800, 3), dtype=np.float32) # Fallback jika file tidak ada
except Exception as e:
    print(f"Error saat memproses gambar dengan PIL: {e}. Menggunakan gambar hitam dummy.")
    image_np = np.zeros((533, 800, 3), dtype=np.float32) # Fallback jika ada error lain

print(f"\n--- Segmentasi Citra Menggunakan K-Means ---")
print(f"Bentuk citra asli (setelah konversi/normalisasi): {image_np.shape}") # (tinggi, lebar, channel_warna)

# Reshape citra menjadi list piksel (setiap piksel adalah instance dengan 3 fitur RGB)
X_image_flat = image_np.reshape(-1, 3) # Ini sekarang akan selalu berfungsi untuk 3 channel

# Latih K-Means pada piksel (berdasarkan warna)
n_colors = 8 # Jumlah cluster warna yang diinginkan
kmeans_colors = KMeans(n_clusters=n_colors, random_state=42, n_init=10)
print(f"Melatih K-Means untuk segmentasi warna dengan {n_colors} warna...")
kmeans_colors.fit(X_image_flat)
print("K-Means segmentasi warna berhasil dilatih.")

# Ganti setiap piksel dengan warna centroid cluster-nya
segmented_img_flat = kmeans_colors.cluster_centers_[kmeans_colors.labels_]
segmented_img = segmented_img_flat.reshape(image_np.shape)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(image_np)
plt.title("Citra Asli")
plt.axis("off")

plt.subplot(122)
plt.imshow(segmented_img)
plt.title(f"Segmentasi K-Means ({n_colors} Warna)")
plt.axis("off")
plt.suptitle("Segmentasi Citra dengan K-Means", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# b. Menggunakan Clustering untuk Preprocessing
# Dataset digit (mirip MNIST)
X_digits, y_digits = load_digits(return_X_y=True)

print("\n--- Clustering untuk Preprocessing (Dataset Digit) ---")
print(f"Bentuk dataset digit: {X_digits.shape}")

# Baseline: Latih Logistic Regression langsung pada data asli
X_train_digits, X_test_digits, y_train_digits, y_test_digits = train_test_split(X_digits, y_digits, random_state=42)

log_reg_baseline = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=2000, random_state=42)
print("Melatih Logistic Regression (baseline)...")
log_reg_baseline.fit(X_train_digits, y_train_digits)
print(f"Akurasi Logistic Regression (baseline): {log_reg_baseline.score(X_test_digits, y_test_digits):.4f}")

# Pipeline: K-Means sebagai preprocessing + Logistic Regression
# K-Means akan mengkluster data dan mengganti fitur asli dengan jarak ke centroid
pipeline_kmeans_logreg = Pipeline([
    ("kmeans", KMeans(n_clusters=50, random_state=42, n_init=10)), # Pilih 50 cluster
    ("log_reg", LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=2000, random_state=42))
])
print("\nMelatih pipeline K-Means + Logistic Regression...")
pipeline_kmeans_logreg.fit(X_train_digits, y_train_digits)
print("Pipeline berhasil dilatih.")
print(f"Akurasi Pipeline K-Means + Logistic Regression: {pipeline_kmeans_logreg.score(X_test_digits, y_test_digits):.4f}")

# Menyetel jumlah cluster (k) untuk preprocessing
param_grid_kmeans = dict(kmeans__n_clusters=range(2, 100)) # Range k dari 2 hingga 99
grid_search_kmeans_pp = GridSearchCV(pipeline_kmeans_logreg, param_grid_kmeans, cv=3, verbose=0, n_jobs=-1)
print("\nMencari jumlah cluster optimal untuk preprocessing dengan GridSearchCV...")
grid_search_kmeans_pp.fit(X_train_digits, y_train_digits)
print(f"Jumlah cluster optimal ditemukan: {grid_search_kmeans_pp.best_params_['kmeans__n_clusters']}")
print(f"Akurasi Pipeline (k optimal): {grid_search_kmeans_pp.score(X_test_digits, y_test_digits):.4f}")


# c. Menggunakan Clustering untuk Semi-Supervised Learning
print("\n--- Clustering untuk Semi-Supervised Learning ---")

n_labeled_digits = 50 # Hanya 50 instance berlabel
# Melatih Logistic Regression hanya dengan data berlabel
log_reg_semisup_baseline = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=2000, random_state=42)
log_reg_semisup_baseline.fit(X_train_digits[:n_labeled_digits], y_train_digits[:n_labeled_digits])
print(f"Akurasi Logistic Regression (50 Label): {log_reg_semisup_baseline.score(X_test_digits, y_test_digits):.4f}")

# Pendekatan Semi-Supervised: Labeling Representative Instances
k_semisup = 50 # Jumlah cluster
kmeans_semisup = KMeans(n_clusters=k_semisup, random_state=42, n_init=10)
kmeans_semisup.fit(X_train_digits)

# Temukan gambar yang paling representatif (terdekat ke centroid)
X_digits_dist = kmeans_semisup.transform(X_train_digits)
representative_digit_idx = np.argmin(X_digits_dist, axis=0)
X_representative_digits = X_train_digits[representative_digit_idx]
y_representative_digits = y_train_digits[representative_digit_idx] # Gunakan label sebenarnya

print("\nMelatih Logistic Regression dengan 50 instance representatif...")
log_reg_representative = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=2000, random_state=42)
log_reg_representative.fit(X_representative_digits, y_representative_digits)
print(f"Akurasi Logistic Regression (50 Representatif): {log_reg_representative.score(X_test_digits, y_test_digits):.4f}")

# Pendekatan Semi-Supervised: Label Propagation
# Propagasi label dari instance representatif ke seluruh cluster
y_train_propagated = np.empty(len(X_train_digits), dtype=np.int64)
for i in range(k_semisup):
    y_train_propagated[kmeans_semisup.labels_ == i] = y_representative_digits[i]

print("\nMelatih Logistic Regression dengan label yang dipropagasi...")
log_reg_propagated = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=2000, random_state=42)
log_reg_propagated.fit(X_train_digits, y_train_propagated)
print(f"Akurasi Logistic Regression (Propagasi Penuh): {log_reg_propagated.score(X_test_digits, y_test_digits):.4f}")

# Propagasi parsial (hanya instance terdekat ke centroid)
percentile_closest = 20
X_cluster_dist = X_digits_dist[np.arange(len(X_train_digits)), kmeans_semisup.labels_]
for i in range(k_semisup):
    in_cluster = (kmeans_semisup.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    # Tandai instance yang lebih jauh dari cutoff
    X_cluster_dist[in_cluster & above_cutoff] = -1

partially_propagated_mask = (X_cluster_dist != -1)
X_train_partially_propagated = X_train_digits[partially_propagated_mask]
y_train_partially_propagated = y_train_propagated[partially_propagated_mask]

print(f"Jumlah instance yang dipropagasi secara parsial: {len(X_train_partially_propagated)}")

print("\nMelatih Logistic Regression dengan label yang dipropagasi secara parsial...")
log_reg_partial_propagated = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=2000, random_state=42)
log_reg_partial_propagated.fit(X_train_partially_propagated, y_train_partially_propagated)
print(f"Akurasi Logistic Regression (Propagasi Parsial): {log_reg_partial_propagated.score(X_test_digits, y_test_digits):.4f}")

# --- 3. DBSCAN ---
# Algoritma berbasis kepadatan

print("\n--- DBSCAN ---")
# Menggunakan dataset moons untuk demonstrasi
X_moons_dbscan, y_moons_dbscan = make_moons(n_samples=1000, noise=0.05, random_state=42)

# Melatih DBSCAN
dbscan = DBSCAN(eps=0.05, min_samples=5) # eps: jarak maksimum antar sampel, min_samples: jumlah sampel dalam radius eps
print("Melatih DBSCAN (eps=0.05, min_samples=5)...")
dbscan.fit(X_moons_dbscan)
print("DBSCAN berhasil dilatih.")

# Label cluster: -1 untuk noise (anomali)
print(f"Contoh label cluster DBSCAN (10 pertama): {dbscan.labels_[:10]}")
print(f"Jumlah instance inti (core instances): {len(dbscan.core_sample_indices_)}")
print(f"Label unik: {np.unique(dbscan.labels_)}")

# Visualisasi DBSCAN (eps=0.05, min_samples=5)
plt.figure(figsize=(8, 6))
plot_clusters(X_moons_dbscan, dbscan.labels_, title="DBSCAN Clusters (eps=0.05)")
plt.show()

# Coba dengan eps lebih besar
dbscan_large_eps = DBSCAN(eps=0.2, min_samples=5)
print("\nMelatih DBSCAN (eps=0.2, min_samples=5)...")
dbscan_large_eps.fit(X_moons_dbscan)
print("DBSCAN (eps=0.2) berhasil dilatih.")
print(f"Label unik (eps=0.2): {np.unique(dbscan_large_eps.labels_)}")

plt.figure(figsize=(8, 6))
plot_clusters(X_moons_dbscan, dbscan_large_eps.labels_, title="DBSCAN Clusters (eps=0.2)")
plt.show()

# DBSCAN tidak memiliki metode predict() secara langsung untuk instance baru.
# Anda bisa melatih classifier terpisah (misal KNeighborsClassifier) pada core instances.
print("\n--- Menggunakan Classifier dengan DBSCAN ---")
# Hanya melatih pada core instances
X_core = dbscan_large_eps.components_
y_core_labels = dbscan_large_eps.labels_[dbscan_large_eps.core_sample_indices_]

knn_dbscan = KNeighborsClassifier(n_neighbors=50)
print("Melatih KNeighborsClassifier pada core instances dari DBSCAN...")
knn_dbscan.fit(X_core, y_core_labels)
print("KNeighborsClassifier berhasil dilatih.")

# Prediksi untuk instance baru
X_new_dbscan = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
new_preds_dbscan = knn_dbscan.predict(X_new_dbscan)
print(f"Prediksi cluster untuk instance baru: {new_preds_dbscan}")

# Mengidentifikasi anomali dengan jarak maksimum
y_dist, y_pred_idx = knn_dbscan.kneighbors(X_new_dbscan, n_neighbors=1)
y_pred_knn_dbscan = dbscan_large_eps.labels_[dbscan_large_eps.core_sample_indices_][y_pred_idx.ravel()]
# Atur ambang batas jarak. Instance yang terlalu jauh dianggap anomali (-1).
# Ambang batas 0.2 adalah contoh, sesuaikan dengan noise dataset.
anom_threshold = 0.2
y_pred_knn_dbscan[y_dist.ravel() > anom_threshold] = -1
print(f"Prediksi cluster (termasuk anomali) untuk instance baru: {y_pred_knn_dbscan}")


# --- 4. Gaussian Mixture Models (GMM) ---
# Model probabilistik yang mengasumsikan data berasal dari campuran distribusi Gaussian.

print("\n--- Gaussian Mixture Models (GMM) ---")
# Membuat dataset dengan 3 blob elips
np.random.seed(42)
X_gmm, y_gmm = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=42)
# Modifikasi data agar sedikit elips dan tumpang tindih untuk menunjukkan kekuatan GMM
X_gmm = X_gmm @ np.array([[0.6, -0.6], [-0.4, 0.8]])

print("Data GMM Sintetis (dengan blob elips):")
plot_clusters(X_gmm, title="Data GMM Sintetis")

# Melatih GMM
n_components_gmm = 3
gmm = GaussianMixture(n_components=n_components_gmm, random_state=42, n_init=10)
print(f"\nMelatih GMM dengan {n_components_gmm} komponen...")
gmm.fit(X_gmm)
print("GMM berhasil dilatih.")

# Parameter yang diestimasi
print(f"Bobot komponen (weights_):\n{gmm.weights_}")
print(f"Means komponen (means_):\n{gmm.means_}")
print(f"Covariances komponen (covariances_):\n{gmm.covariances_}")
print(f"Model converged: {gmm.converged_}")
print(f"Jumlah iterasi: {gmm.n_iter_}")

# Prediksi cluster (hard clustering)
y_pred_gmm = gmm.predict(X_gmm)
print(f"Contoh label cluster GMM (5 pertama): {y_pred_gmm[:5]}")

# Prediksi probabilitas (soft clustering)
y_proba_gmm = gmm.predict_proba(X_gmm)
print(f"Contoh probabilitas cluster GMM (5 pertama):\n{y_proba_gmm[:5]}")

# Mengestimasi kepadatan (score_samples)
log_densities = gmm.score_samples(X_gmm)
print(f"Contoh log densitas (5 pertama): {log_densities[:5]}")

# Memvisualisasikan batas keputusan dan kontur densitas GMM
def plot_gmm_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True, show_xlabels=True, show_ylabels=True, title=""):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    x1, x2 = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    X_plot = np.c_[x1.ravel(), x2.ravel()]

    log_densities = clusterer.score_samples(X_plot)
    densities = np.exp(log_densities).reshape(x1.shape)

    y_pred = clusterer.predict(X_plot).reshape(x1.shape)

    plt.contourf(x1, x2, y_pred, cmap="Pastel2")
    plt.contour(x1, x2, y_pred, linewidths=1, colors='k')

    plt.contour(x1, x2, densities, cmap="viridis", linewidths=2)

    if show_centroids:
        plot_centroids(clusterer.means_, clusterer.weights_)

    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title(title)

plt.figure(figsize=(10, 8))
plot_gmm_decision_boundaries(gmm, X_gmm, title="GMM Decision Boundaries dan Density Contours")
plt.show()

# b. Anomaly Detection Menggunakan GMM
print("\n--- Anomaly Detection Menggunakan GMM ---")
# Ambil 4% instance dengan densitas terendah sebagai anomali
density_threshold = np.percentile(log_densities, 4)
anomalies = X_gmm[log_densities < density_threshold]

plt.figure(figsize=(10, 8))
plot_gmm_decision_boundaries(gmm, X_gmm, title="GMM Anomaly Detection")
plt.scatter(anomalies[:, 0], anomalies[:, 1], color='r', marker='x', s=100, label="Anomali")
plt.legend()
plt.show()


# c. Memilih Jumlah Cluster (Komponen) untuk GMM
# Menggunakan BIC (Bayesian Information Criterion) dan AIC (Akaike Information Criterion)
print("\n--- Memilih Jumlah Cluster Optimal untuk GMM (BIC & AIC) ---")
n_components_range = range(1, 10)
bics = []
aics = []

for n_comp in n_components_range:
    gmm_comp = GaussianMixture(n_components=n_comp, random_state=42, n_init=10)
    gmm_comp.fit(X_gmm)
    bics.append(gmm_comp.bic(X_gmm))
    aics.append(gmm_comp.aic(X_gmm))

plt.figure(figsize=(10, 5))
plt.plot(n_components_range, bics, "bo-", label="BIC")
plt.plot(n_components_range, aics, "ro-", label="AIC")
plt.xlabel("Jumlah Komponen")
plt.ylabel("Kriteria Informasi")
plt.title("BIC dan AIC vs. Jumlah Komponen GMM")
plt.legend()
plt.grid(True)
plt.show()

print(f"Jumlah komponen optimal berdasarkan BIC/AIC (nilai terendah): {n_components_range[np.argmin(bics)]} (BIC) atau {n_components_range[np.argmin(aics)]} (AIC)")


# d. Bayesian Gaussian Mixture Models (Otomatis memilih jumlah komponen)
print("\n--- Bayesian Gaussian Mixture Models (BGMM) ---")
# Set n_components ke nilai yang lebih tinggi dari yang diharapkan. BGMM akan membuat bobot nol untuk komponen yang tidak perlu.
bgm = BayesianGaussianMixture(n_components=10, random_state=42, n_init=10)
print("Melatih Bayesian Gaussian Mixture Model...")
bgm.fit(X_gmm)
print("Bayesian GMM berhasil dilatih.")

print(f"Bobot komponen yang ditemukan:\n{np.round(bgm.weights_, 2)}")
print("Perhatikan komponen dengan bobot mendekati nol diabaikan.")

plt.figure(figsize=(10, 8))
plot_gmm_decision_boundaries(bgm, X_gmm, title="Bayesian GMM Decision Boundaries dan Density Contours")
plt.show()

print("\n--- Selesai Reproduksi Kode Chapter 9 ---")
```
