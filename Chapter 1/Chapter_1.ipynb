```python
import matplotlib.pyplot as pltg
import numpy as np
import pandas as pd
import sklearn.linear_model

# -----------------------------------------------------------------------------
# Bagian 1: Memuat dan Menyiapkan Data (Pengalaman E)
# -----------------------------------------------------------------------------
# Dalam Bab 1, buku ini menggunakan contoh kepuasan hidup vs. PDB per kapita.
# Kita akan mereplikasi ini dengan data sederhana yang serupa.
# Asumsi: Anda memiliki file CSV ini di direktori yang sama atau path yang ditentukan.
# Di buku, data diambil dari OECD dan IMF. Untuk tujuan simulasi,
# kita akan menggunakan data yang disederhanakan.

# Contoh data (biasanya dari file CSV seperti yang disebutkan di buku)
# Data ini mencerminkan tren yang diamati di Bab 1.
# Misalkan PDB per kapita (dalam ribuan USD) dan Kepuasan Hidup (skala 0-10).
data_sederhana = {
    'GDP per capita': [10, 20, 30, 40, 50, 60, 70, 80],
    'Life satisfaction': [4.5, 5.5, 6.2, 6.8, 7.3, 7.5, 7.8, 8.0]
}
df = pd.DataFrame(data_sederhana)

# Pisahkan fitur (prediktor) dan label (target)
X = df[['GDP per capita']].values # Fitur harus 2D array
y = df['Life satisfaction'].values # Label bisa 1D array

print("Data yang digunakan:")
print(df)
print("\nBentuk fitur (X):", X.shape)
print("Bentuk label (y):", y.shape)

# -----------------------------------------------------------------------------
# Bagian 2: Visualisasi Data (Membantu Pemahaman)
# -----------------------------------------------------------------------------
# Visualisasi adalah langkah penting untuk mendapatkan wawasan (seperti yang dibahas
# di Bab 1 tentang "Discover and Visualize the Data to Gain Insights").

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Pelatihan')
plt.xlabel('PDB per Kapita (dalam ribuan USD)')
plt.ylabel('Kepuasan Hidup')
plt.title('PDB per Kapita vs. Kepuasan Hidup')
plt.grid(True)
plt.legend()
plt.show()

# -----------------------------------------------------------------------------
# Bagian 3: Memilih Model dan Melatihnya
# -----------------------------------------------------------------------------
# Seperti yang dijelaskan dalam Bab 1, "model-based learning" melibatkan pemilihan
# model (misal: Regresi Linier) dan melatihnya dengan data.

# Pilih model regresi linier (seperti yang dilakukan Aurélien Géron di Bab 1)
model_regresi_linier = sklearn.linear_model.LinearRegression()

# Latih model dengan data (belajar dari pengalaman E)
# Model akan menemukan parameter (koefisien dan intercept) terbaik
# yang sesuai dengan data.
print("\nMelatih model Regresi Linier...")
model_regresi_linier.fit(X, y)
print("Model berhasil dilatih.")

# Tampilkan parameter model yang ditemukan
print(f"Intercept (theta_0): {model_regresi_linier.intercept_:.2f}")
print(f"Koefisien (theta_1): {model_regresi_linier.coef_[0]:.2f}")

# -----------------------------------------------------------------------------
# Bagian 4: Membuat Prediksi (Tugas T)
# -----------------------------------------------------------------------------
# Setelah model dilatih, kita bisa menggunakannya untuk membuat prediksi pada
# instance baru (inferensi).

# Misalkan kita ingin memprediksi kepuasan hidup untuk PDB per kapita 45 ribu USD.
new_gdp = np.array([[45]]) # Fitur baru harus 2D array
predicted_satisfaction = model_regresi_linier.predict(new_gdp)

print(f"\nPrediksi Kepuasan Hidup untuk PDB per Kapita 45 ribu USD: {predicted_satisfaction[0]:.2f}")

# -----------------------------------------------------------------------------
# Bagian 5: Visualisasi Garis Regresi (Kinerja Model)
# -----------------------------------------------------------------------------
# Visualisasikan model yang dilatih pada scatter plot.

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Pelatihan Asli')
plt.plot(X, model_regresi_linier.predict(X), color='red', linestyle='-', label='Garis Regresi Linier')
plt.scatter(new_gdp, predicted_satisfaction, color='green', marker='X', s=200, label='Prediksi Baru')
plt.xlabel('PDB per Kapita (dalam ribuan USD)')
plt.ylabel('Kepuasan Hidup')
plt.title('Model Regresi Linier Dilatih dan Prediksi')
plt.grid(True)
plt.legend()
plt.show()

print("\n--- Selesai Simulasi Chapter 1 ---")
print("Ini menunjukkan langkah-langkah dasar Pembelajaran Mesin yang diperkenalkan")
print("dalam Bab 1, meskipun bab tersebut sebagian besar bersifat teoritis.")
```
